# Preprocessing


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns #Seaborn is a Python data visualization library based on matplotlib
import datetime as dt
```

```python
hotel_data = pd.read_csv('hotelsBookingDemand.csv')
```

```python
hotel_data
```

###  Checking missing value

```python
hotel_data.isna().sum()
```

```python
#fill missing value in country and company cloumn with NULL
hotel_data.fillna({'company':0,'agent':0,})
```

```python
#drop rows with missingvalues
hotel_data.dropna(axis=0, how='any', inplace=True)
```

```python
#let's check if still missing values
hotel_data.isna().sum().sum()
```

```python
#convert to arrival date 
arrival_date=pd.to_datetime(hotel_data['arrival_date_year'].astype(str)  + hotel_data['arrival_date_month']+hotel_data['arrival_date_day_of_month'].astype(str), format='%Y%B%d')
hotel_data['arrival_date']=arrival_date
hotel_data.drop(['arrival_date_year','arrival_date_day_of_month','arrival_date_month'],axis=1,inplace=True)
```

```python

hotel_data['reservation_status_date']=pd.to_datetime(hotel_data['reservation_status_date'])
```

```python
# that the timestamp of the variable reservation_status_date must occur after or at the same date as the input variable arrival_date
d=hotel_data.iloc[np.where(hotel_data['reservation_status_date']<hotel_data['arrival_date'])]
d
```

# Additional preprocessing


```python
X=hotel_data.loc[:, hotel_data.columns != 'is_canceled']
y= hotel_data['is_canceled']
```

### Encodage des variables catÃ©goriques

```python
from sklearn.preprocessing import OneHotEncoder
categorical_features=hotel_data.select_dtypes(include=['object']).columns.tolist()
oneHotEncoder=OneHotEncoder(handle_unknown = "ignore")
```

### Normalisation 

```python
from sklearn.preprocessing import StandardScaler
numerical_features=['total_of_special_requests','required_car_parking_spaces','adr','days_in_waiting_list','company','agent',
                    'booking_changes','previous_bookings_not_canceled','previous_cancellations','is_repeated_guest','babies',
                    'children','adults','stays_in_week_nights','arrival_date_week_number','lead_time']
standardScaler=StandardScaler()
```

### PCA

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=10)
```

## Pipeline

```python
from sklearn.pipeline import make_pipeline
from sklearn.pipeline import Pipeline
from sklearn.compose import make_column_selector, make_column_transformer
preprocessor=make_column_transformer((oneHotEncoder,categorical_features),
                                        (standardScaler,numerical_features))
```

```python
model_pipeline=Pipeline(steps=[("preprocessor",preprocessor),("pca",pca)])
print (f'Fitting PCA ...')
model_pipeline.fit_transform(X,y)
```

# Modeling

```python
from sklearn.model_selection import train_test_split
```

```python
#splitting data into testing and training
i=20/100
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=i,random_state=0)
```

#### Random forest

```python
from sklearn.ensemble import RandomForestClassifier
model= RandomForestClassifier(max_depth=7)
```

```python
model_pipeline=make_pipeline(preprocessor,pca,model)
print (f'Fitting RandomForestClassifier ...')
model_pipeline.fit(X_train,y_train)
print(f'scoring RandomForestClassifier  :', model_pipeline.score(X_test,y_test))
```

```python
y_predR = model_pipeline.predict(X_test)

y_predR
```

#### Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
modelLg= LogisticRegression()
```

```python
model_pipelineLg=Pipeline(steps=[("preprocessor",preprocessor),("pca",pca),('model',modelLg)])
print (f'Fitting LOGISTICREGRESSION ...')
model_pipelineLg.fit(X_train,y_train)
print(f'scoring LOGISTICREGRESSION  :', model_pipelineLg.score(X_test,y_test))
```

```python
y_pred = model_pipelineLg.predict(X_test)

y_pred
```

# comparison
